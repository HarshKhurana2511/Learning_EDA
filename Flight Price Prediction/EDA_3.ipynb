# Feature Engineering


```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], dayfirst = True)
df.head()
```

```python
df["Date"] = df['Date_of_Journey'].dt.day
df["Month"] = df['Date_of_Journey'].dt.month
df["Year"] = df['Date_of_Journey'].dt.year
df.head()
```

```python
df.info()
```

```python
df = df.drop("Date_of_Journey", axis=1)
df.head()
```

```python
df.Arrival_Time.str.split(' ').str[0]
```

```python
df.Arrival_Time = df.Arrival_Time.str.split(' ').str[0]
```

```python
df.head()
```

```python
df["Arrival_Hour"] = df.Arrival_Time.str.split(':').str[0].astype(int)
df["Arrival_Min"] = df.Arrival_Time.str.split(':').str[1].astype(int)
df = df.drop("Arrival_Time", axis=1)
df.head()
```

```python
df["Dep_Hour"] = df.Dep_Time.str.split(':').str[0].astype(int)
df["Dep_Min"] = df.Dep_Time.str.split(':').str[1].astype(int)
df = df.drop("Dep_Time", axis=1)
df.head()
```

```python
df.info()
```

```python
df["Total_Stops"].unique()
```

```python
df.Total_Stops = df["Total_Stops"].map({
    'non-stop' : 0,
    '1 stop' : 1,
    '2 stops' : 2,
    '3 stops' : 3,
    '4 stops' : 4
})
```

```python
df.isnull().sum()
```

```python
df[df.Total_Stops.isnull()]
```

```python
df = df.dropna(subset = ['Total_Stops']) #Drops the row with the Nan value in column "Total_Stops"
df.info()
```

```python
df = df.drop("Route", axis=1)
```

```python
df.head()
```

```python
#Convert total time in minutes
df["Duration_Hour"] = df.Duration.str.split(' ').str[0].str.split('h').str[0]
df.head()
```

```python
df[df.Duration_Hour == '5m']
```

```python
df = df.drop([6474, 13343], axis=0)
```

```python
df[df.Duration_Hour == '5m']
```

```python
df.info()
```

```python
df["Duration_Hour"] = df["Duration_Hour"].astype(int)*60
```

```python
df["Duration_Min"] = df.Duration.str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df.Duration_Min.isnull().sum()
```

```python
df['Duration_Min'] = df.Duration_Min.fillna(0)
df.Duration_Min.isnull().sum()
```

```python
df['Duration_Min'] = df['Duration_Min'].astype(int)
df.head()
```

```python
df.info()
```

```python
df.Duration = df.Duration_Hour + df.Duration_Min
df.head()
```

```python
df = df.drop(["Duration_Hour", "Duration_Min"], axis=1)
df.info()
```
